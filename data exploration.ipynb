<font face="微软雅黑" size=5>3.1. Summary Statistics

<font face="微软雅黑" size=4>Summary statistics are quantities, such as the mean and standard deviation, that capture various characteristics of a potentially large set of values with a single number or a small set of numbers. In this tutorial, we will use the Iris sample data, which contains information on 150 Iris flowers, 50 each from one of three Iris species: Setosa, Versicolour, and Virginica. Each flower is characterized by five attributes:

<font face="微软雅黑" size=4>sepal length in centimeters  
sepal width in centimeters  
petal length in centimeters  
petal width in centimeters  
class (Setosa, Versicolour, Virginica)

<font face="微软雅黑" size=4>In this tutorial, you will learn how to:  
      Load a CSV data file into a Pandas DataFrame object.  
      Compute various summary statistics from the DataFrame.

<font face="微软雅黑" size=4>To execute the sample program shown here, make sure you have installed the Pandas library (see Module 2).

<font face="微软雅黑" size=4>1. First, you need to download the Iris dataset from the UCI machine learning repository.

<font face="微软雅黑" size=4>Code: The following code uses Pandas to read the CSV file and store them in a DataFrame object named data. Next, it will display the first five rows of the data frame.

```python
import pandas as pd

data = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',header=None)
data.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'class']

data.head()
```

<font face="微软雅黑" size=4>2. For each quantitative attribute, calculate its average, standard deviation, minimum, and maximum values.

Code:

```python
from pandas.api.types import is_numeric_dtype

for col in data.columns:
    if is_numeric_dtype(data[col]):
        print('%s:' % (col))
        print('\t Mean = %.2f' % data[col].mean())
        print('\t Standard deviation = %.2f' % data[col].std())
        print('\t Minimum = %.2f' % data[col].min())
        print('\t Maximum = %.2f' % data[col].max())
```

<font face="微软雅黑" size=4>3. For each pair of attributes, we can use a scatter plot to visualize their joint distribution.  
  
Code:

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(3, 2, figsize=(12,12))
index = 0
for i in range(3):
    for j in range(i+1,4):
        ax1 = int(index/2)
        ax2 = index % 2
        axes[ax1][ax2].scatter(data[data.columns[i]], data[data.columns[j]], color='red')
        axes[ax1][ax2].set_xlabel(data.columns[i])
        axes[ax1][ax2].set_ylabel(data.columns[j])
        index = index + 1
```

```python
import matplotlib.pyplot as plt
fig,axes = plt.subplots(3,2,figsize=(12,12))
index=0
for i in range(3):
    for j in range(i+1,4):
        x=int(index/2)
        y=index%2
        axes[x][y].scatter(data1[data1.columns[i]],data1[data1.columns[j]],color = 'yellow')
        axes[x][y].set_xlabel(data1.columns[i])
        axes[x][y].set_ylabel(data1.columns[j])
        index= index + 1
```

<font face="微软雅黑" size=4>4. Parallel coordinates can be used to display all the data points simultaneously. Parallel coordinates have one coordinate axis for each attribute, but the different axes are parallel to one other instead of perpendicular, as is traditional. Furthermore, an object is represented as a line instead of as a point. In the example below, the distribution of values for each class can be identified in a separate color.  
  
Code:

```python
from pandas.plotting import parallel_coordinates
%matplotlib inline

parallel_coordinates(data, 'class')
```
